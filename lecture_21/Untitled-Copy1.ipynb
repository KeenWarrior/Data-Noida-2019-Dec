```python
# custom filter size
f = 20
fil = np.zeros((f, f))
```

```python
# custom value for filter
fil[:, :] = 1
```

```python
red = pan_np[:, :, 0]
```

```python
padded = np.pad(red, f // 2)
```

```python
result = np.zeros_like(red)
```

```python
for r in range(red.shape[0]):
    for c in range(red.shape[1]):
        cut = padded[r:r+f, c:c+f]
        result[r, c] = np.abs((fil * cut).mean())
```

```python
plt.imshow(result, cmap="gray")
```
