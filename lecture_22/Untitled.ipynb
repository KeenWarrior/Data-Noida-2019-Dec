```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing.image import load_img
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = np.array(load_img("../datasets/boxes.jpg", ))/255
```

```python
img.shape
```

```python
gen = ImageDataGenerator(horizontal_flip=True, zoom_range=1.5, rotation_range=30)
```

```python
flow = gen.flow(np.array([img]))
```

```python
for im in flow.next():
    plt.imshow(im)
```
